```python
# Csak a nem-NaN értékeket tartjuk meg
from itertools import product
guides = cards["Guide"].dropna().unique()
targets = cards["Tárgya"].dropna().unique()

# Összes kombináció: minden utasítás minden tárggyal
combinations = [f"{g} {t}" for g, t in product(guides, targets)]

# Eredmény DataFrame
combi_df = pd.DataFrame(combinations, columns=["Guide x Tárgya kombináció"])

# Megnézünk pár sort
print(combi_df.head(10))
```
